```python
import numpy as np
import matplotlib.pyplot as pp
import math
from scipy.interpolate import interp1d
```

```python
list=[[1,1,1,0,0,0]]
t1=[10,400]
for j in t1:
    y=[]
    for i in range(6):
        try:
            t=1/(1+(math.exp((i-2.4)/(1.3806452*math.pow(10,-23)*j))))
        except OverflowError :
            t=0
        y.append(t)
    list.append(y)
        
```

```python
e=[0,1,2,3,4,5]
list
```

```python
pp.plot([2.4,2.4,2.4,2.4],[0,0,1,1],label="F(E)")
pp.plot([0,1,2,3,4,5],[0.5,0.5,0.5,0.5,0.5,0.5],'o:b',label="0.5")
pp.plot()
x=np.array(e)
x1,x2,x3=np.array(list[0]),np.array(list[1]),np.array(list[2])
cubic=interp1d(x,x1,kind="cubic")
X=np.linspace(0,x.max(),500)
y2=cubic(X)
pp.plot(X,y2,label="0k")
cubic=interp1d(x,x2,kind="cubic")
X=np.linspace(x.min(),x.max(),500)
y=cubic(X)
pp.plot(X,y,label="10k",linestyle="dotted",linewidth='3')
cubic=interp1d(x,x3,kind="cubic")
X=np.linspace(x.min(),x.max(),500)
y=cubic(X)
pp.plot(X,y,label="400k",linestyle='--',linewidth='2.5')
pp.xlabel("E->")
pp.ylabel("F(E)")
pp.title("Fermi Dirac Dist.")
pp.legend()
pp.show()
```
